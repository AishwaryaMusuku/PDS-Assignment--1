```python
import pandas as pd

# importing data
data = pd.read_csv('/content/data_clean.csv')
data.head()
```

```python
#selecting the specific columns
selected_data = data[['Name', 'Location', 'Year','Price']]
selected_data.head()
```

```python
#Filtering dataset with location and Owner_Type
filtered_data = data[(data["Location"] == "Pune") & (data["Owner_Type"] == "First")]
filtered_data.head()
```

```python
#Renaming columns Location, Mileage, Engine, Power with Place, Mileage in kmpl, Engine in CC, Power in bhp
renamed_columns = data.rename(columns={"Location": "Place","Mileage": "Mileage in kmpl", "Engine": "Engine in CC", "Power": "Power in bhp"})
renamed_columns.head()
```

```python
# Mutate operation
# Adding new column "Kilometers_Per_Year"
current_year = 2024
data['Age'] = current_year - data['Year']
data['Kilometers_Per_Year'] = data['Kilometers_Driven'] / data['Age']
data.head()
```

```python
# Arranging data based on Car Year
arranged_data = data.sort_values(by="Year")
arranged_data.head()
```

```python
# Group By operation
# Sum of Kilometers_Driven and Mean price for each year in each location
summary_data = data.groupby(["Year", "Location"]).agg({"Kilometers_Driven": "sum","Price": "mean"})
summary_data.head(20)
```
