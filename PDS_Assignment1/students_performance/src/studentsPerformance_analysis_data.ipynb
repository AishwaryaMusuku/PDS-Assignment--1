```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# importing studentsPerformance data
studentsPerformance_data_raw = pd.read_csv('/content/students_performance/data_raw/StudentsPerformance.csv')

#Replacing the categorical values('female', 'male') with numerical values('0', '1')
studentsPerformance_data_raw['gender'].replace(['female', 'male'], [0, 1], inplace=True)

#Replacing the categorical values('none', 'completed') with numerical values('0', '1')
studentsPerformance_data_raw['test preparation course'].replace(['none', 'completed'], [0, 1], inplace=True)

#dropping the rows with no values or having na
studentsPerformance_raw_data = studentsPerformance_data_raw.dropna()

# Removed the duplicate rows
studentsPerformance_data_raw = studentsPerformance_data_raw.drop_duplicates()

studentPerformance_data_clean = studentsPerformance_data_raw

# Visualization 1: Box plot of reading score distribution by lunch type
plt.figure(figsize=(8, 6))
sns.boxplot(x='lunch', y='reading score', data=studentPerformance_data_clean, hue='lunch', dodge=False)
plt.title('Reading Score Distribution by Lunch Type')
plt.xlabel('Lunch Type')
plt.ylabel('Reading Score')
plt.savefig('/content/students_performance/results/readingscores_lunchType.png')
plt.show()
```

```python
# Visualization 2: Bar plot of average math scores by race/ethnicity
plt.figure(figsize=(10, 6))
sns.barplot(x='race/ethnicity', y='math score', data=studentPerformance_data_clean, errorbar=None)
plt.title('Average Math Scores by Race/Ethnicity')
plt.xlabel('Race/Ethnicity')
plt.ylabel('Average Math Score')
plt.show()
```

```python
#Visualisation 3: Gender Distribution
plt.bar(studentPerformance_data_clean['gender'].value_counts().index, studentPerformance_data_clean['gender'].value_counts().values, color = ['green', 'blue'])
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(studentPerformance_data_clean['gender'].value_counts().index,['Female','Male'])
```

```python
#visualisation 4: Frequency distribution of math, reading and writing scores

plt.figure(figsize=(14, 5))

plt.subplot(1, 3, 1)
sns.histplot(studentPerformance_data_clean['math score'],label='math score')
plt.title('Math Score distribution')
plt.subplot(1, 3, 2)
sns.histplot(studentPerformance_data_clean['reading score'],label='reading score')
plt.title('Reading Score distribution')
plt.subplot(1, 3, 3)
sns.histplot(studentPerformance_data_clean['writing score'],label='writing score')
plt.title('Writing Score distribution')
plt.show()
```

```python
# Visualization 5: Scatter plot of math vs reading scores by gender
plt.figure(figsize=(8, 6))
sns.scatterplot(x='math score', y='reading score', hue='gender', data=studentPerformance_data_clean)
plt.title('Math vs. Reading Scores by Gender')
plt.xlabel('Math Score')
plt.ylabel('Reading Score')
plt.show()
```
