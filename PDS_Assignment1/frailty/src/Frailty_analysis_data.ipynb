```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# importing frailty data
frailty_data_raw = pd.read_csv('/content/frailty/data_raw/frailty_raw_data.csv')

#Updated the column names
frailty_data_raw.columns = ['Height', 'Weight', 'Age', 'Grip_Strength', 'Frailty']

# Removed the duplicate rows
frailty_data_raw = frailty_data_raw.drop_duplicates()

#Replaced the categorical values of Frailty column with numerical values
frailty_data_raw['Frailty'].replace(['Y', 'N'], [0,1], inplace=True)

frailty_data_clean = frailty_data_raw

# Plotting the relationship between Weight and Frailty
plt.figure(figsize=(10, 6))
sns.boxplot(x='Frailty', y='Weight', data=frailty_data_clean, hue='Frailty', dodge=False)
plt.title('Relationship between Weight and Frailty')
plt.xlabel('Frailty')
plt.ylabel('Weight')
plt.savefig('/content/frailty/results/relation_between_weight_frailty.png')
plt.show()
```

```python
# Plotting the relationship between Age and Frailty
plt.figure(figsize=(10, 6))
sns.boxplot(x='Frailty', y='Age', data=frailty_data_clean, hue='Frailty', dodge=False)
plt.title('Relationship between Age and Frailty')
plt.xlabel('Frailty')
plt.ylabel('Age')
plt.savefig('/content/frailty/results/relation_between_age_frailty.png')
plt.show()
```
